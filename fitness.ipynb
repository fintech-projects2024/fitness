```python
import tkinter as tk
import mysql.connector
from tkinter import messagebox,PhotoImage

# Defined global variables
date_entry = None
exercise_entry = None
duration_entry = None
calories_entry = None

#mysql connection
conn = mysql.connector.connect(
            host="localhost",
            user="root",
            password="DPPandya_19",
            database="fitness_db"
        )
cursor = conn.cursor()

#WORKOUT
def add_workout():
    global date_entry, exercise_entry, duration_entry, calories_entry,conn, cursor

    # Getting input values
    date = date_entry.get()
    exercise = exercise_entry.get()
    duration = int(duration_entry.get())
    calories_burned = int(calories_entry.get())

    # Insert the workout data into the database
    cursor.execute("INSERT INTO workouts (date, exercise, duration, calories_burned) VALUES (%s, %s, %s, %s)",
                   (date, exercise, duration, calories_burned))
    conn.commit()
    
    date_entry.delete(0, tk.END)
    exercise_entry.delete(0, tk.END)
    duration_entry.delete(0, tk.END)
    calories_entry.delete(0, tk.END)

def open_workout():
    global date_entry, exercise_entry, duration_entry, calories_entry
    root.withdraw()

    # Create a new workout frame
    workout_frame = tk.Toplevel(root)
    workout_frame.title("Fitness Tracker")
    workout_frame.geometry("2000x1000")
    workout_frame.configure(bg='#D3D3D3')
    custom_font = ("Helvetica", 20)
    # custom_font2 = ("Helvetica", 40)

    # Labels and Entry Widgets
   
    date_label = tk.Label(workout_frame, text="Date (YYYY-MM-DD):",font=custom_font)
    date_label.pack()
    date_entry = tk.Entry(workout_frame,font=custom_font)
    date_entry.pack()

    exercise_label = tk.Label(workout_frame, text="Exercise:",font=custom_font)
    exercise_label.pack()
    exercise_entry = tk.Entry(workout_frame,font=custom_font)
    exercise_entry.pack()

    duration_label = tk.Label(workout_frame, text="Duration (minutes):",font=custom_font)
    duration_label.pack()
    duration_entry = tk.Entry(workout_frame,font=custom_font)
    duration_entry.pack()

    calories_label = tk.Label(workout_frame, text="Calories Burned:",font=custom_font)
    calories_label.pack()
    calories_entry = tk.Entry(workout_frame,font=custom_font)
    calories_entry.pack()

    # Add Workout Button
    add_button = tk.Button(workout_frame, text="Add Workout",font=custom_font, command=add_workout, bg='#007BFF')
    add_button.pack()

    back_button = tk.Button(workout_frame, text="Back to Home",font=custom_font, command=lambda: back_to_workout(workout_frame), bg='lightgreen')
    back_button.pack()

#HISTORY
def open_history():
    global root, date_entry, conn, cursor  
    root.withdraw()
  

    def show_history():
        selected_date = date_entry.get()  
        cursor.execute("SELECT date, exercise, duration, calories_burned FROM workouts WHERE date = %s", (selected_date,))
        workout_history = cursor.fetchall()

        conn.close()

        history_label = tk.Label(history_frame, text="Workout History for " + selected_date)
        history_label.pack()

        text_widget = tk.Text(history_frame, height=10, width=40, font="calibri, 20")
        text_widget.pack()

        if workout_history:
            for entry in workout_history:
                text_widget.insert(tk.END, f"Date: {entry[0]}\nExercise: {entry[1]}\nDuration: {entry[2]} minutes\nCalories Burned: {entry[3]}\n\n")
        else:
            text_widget.insert(tk.END, "No workout history for the specified date.")

    history_frame = tk.Toplevel(root)
    history_frame.title("Workout History")
    history_frame.geometry("2000x1000")
    custom_font = ("Helvetica", 20)

    date_label = tk.Label(history_frame, text="Date (YYYY-MM-DD):", font=custom_font)
    date_label.pack()
    date_entry = tk.Entry(history_frame, font=custom_font)
    date_entry.pack()

    show_button = tk.Button(history_frame, text="Show", command=show_history, font=custom_font,bg='#007BFF')
    show_button.pack()

    back_button = tk.Button(history_frame, text="Back to Home", font=custom_font, command=lambda: back_to_workout(history_frame),bg='lightgreen')
    back_button.pack()
    
#DIET
def open_diet_suggestion():
    global root
    root.withdraw()

    diet_frame = tk.Toplevel(root)
    diet_frame.title("Diet Suggestion")
    diet_frame.geometry("2000x1000")
    custom_font = ("Helvetica", 20)

   # suggestions for the diet
    diet_label = tk.Label(diet_frame, text="Diet Suggestion:", font=custom_font, bg='#007BFF')
    diet_label.pack()

    diet_text = tk.Text(diet_frame, height=10, width=40, font=custom_font)
    diet_text.pack()

    diet_text.insert(tk.END, "Breakfast:\n- Oatmeal with fruits\n- Boiled eggs\n\n")
    diet_text.insert(tk.END, "Lunch:\n- Grilled chicken salad\n- Quinoa\n\n")

    diet_text.insert(tk.END, "Dinner:\n- Baked salmon\n- Steamed vegetables\n\n")

    back_button = tk.Button(diet_frame, text="Back to Home", font=custom_font, command=lambda: back_to_workout(diet_frame),bg='lightgreen')
    back_button.pack()
    

#BMI
def open_bmi_calculator():
    root.withdraw()
  
    def reset_entry():
        age_entry.delete(0, 'end')
        height_entry.delete(0, 'end')
        weight_entry.delete(0, 'end')

    def calculate_bmi():
        kg = int(weight_entry.get())
        m = int(height_entry.get()) / 100
        bmi = kg / (m * m)
        bmi = round(bmi, 1)
        bmi_index(bmi)

    def bmi_index(bmi):
        if bmi < 18.5:
            messagebox.showinfo('BMI Calculator', f'BMI = {bmi} is Underweight')
        elif (bmi >= 18.5) and (bmi < 24.9):
            messagebox.showinfo('BMI Calculator', f'BMI = {bmi} is Normal')
        elif (bmi >= 24.9) and (bmi < 29.9):
            messagebox.showinfo('BMI Calculator', f'BMI = {bmi} is Overweight')
        elif (bmi >= 29.9):
            messagebox.showinfo('BMI Calculator', f'BMI = {bmi} is Obesity')
        else:
            messagebox.showerror('BMI Calculator', 'Something went wrong!')

    calc = tk.Toplevel(root)
    calc.title('BMI Calculator')
    calc.geometry('2000x1000')
    custom_font = ("Helvetica", 20)
    
    frame = tk.Frame(calc, padx=10, pady=10)
    frame.pack(expand=True)

    age_label = tk.Label(frame, text="Enter Age", font=custom_font)
    age_label.grid(row=1, column=1)
    age_entry = tk.Entry(frame, font=custom_font)
    age_entry.grid(row=1, column=2, pady=5)

    var = tk.StringVar()  

    gen_label = tk.Label(frame, text='Select Gender', font=custom_font)
    gen_label.grid(row=2, column=1)

    frame2 = tk.Frame(frame)
    frame2.grid(row=2, column=2, pady=5)

    male_rb = tk.Radiobutton(frame2, text='Male', variable=var, value=1, font=custom_font)
    male_rb.pack(side='top')  
    female_rb = tk.Radiobutton(frame2, text='Female', variable=var, value=2, font=custom_font)
    female_rb.pack(side='top')  

    height_lb = tk.Label(frame, text="Enter Height (cm)  ", font=custom_font)
    height_lb.grid(row=3, column=1)
    height_entry = tk.Entry(frame, font=custom_font)
    height_entry.grid(row=3, column=2, pady=5)

    weight_lb = tk.Label(frame, text="Enter Weight (kg)  ", font=custom_font)
    weight_lb.grid(row=4, column=1)
    weight_entry = tk.Entry(frame, font=custom_font)
    weight_entry.grid(row=4, column=2, pady=5)

    frame3 = tk.Frame(frame)
    frame3.grid(row=5, columnspan=3, pady=10)

    cal_btn = tk.Button(frame3, text='Calculate', font=custom_font, command=calculate_bmi,bg='#007BFF')
    cal_btn.pack(side='left')

    reset_btn = tk.Button(frame3, text='Reset', font=custom_font, command=reset_entry, bg='red')
    reset_btn.pack(side='left')

    exit_btn = tk.Button(frame3, text='Back to Home', font=custom_font, command=lambda: back_to_workout(calc), bg='lightgreen')
    exit_btn.pack(side='right')

    def back_to_workout(calc):
        calc.destroy()
        root.deiconify() 
    
#BACK TO HOME
def back_to_workout(diet_frame):
    diet_frame.destroy()
    root.deiconify()

root = tk.Tk()
root.title("Fitness Tracker")
root.geometry("2000x1000")
bg_image = PhotoImage(file="bg.png")

# main frame
main_frame = tk.Frame(root)
main_frame.pack(fill='both',expand=True)
custom_font = ("Helvetica", 20)
custom_font2 = ("Helvetica", 40)

bg_label = tk.Label(main_frame, image=bg_image)
bg_label.place(relwidth=1, relheight=1) 

# buttons for Workout, History, BMI Calculator, and Diet
workout_button = tk.Button(main_frame, text="Workout", command=open_workout,bg='grey',font=custom_font)
history_button = tk.Button(main_frame, text="History", command=open_history, bg='grey',font=custom_font)
bmi_button = tk.Button(main_frame, text="BMI Calculator", command=open_bmi_calculator, bg='grey',font=custom_font)
diet_button = tk.Button(main_frame, text="Diet", command=open_diet_suggestion, bg='grey',font=custom_font)
quote_label = tk.Label(main_frame,text= "It's not about just calories, ",font=custom_font2)
quote_label2 = tk.Label(main_frame,text= "It's about feeling better",font=custom_font2)

workout_button.place(x=300, y=70)
history_button.place(x=500, y=70)
bmi_button.place(x=650, y=70)
diet_button.place(x=900, y=70)
quote_label.place(x=330, y= 500)
quote_label2.place(x=360, y= 570)
root.mainloop()
```
